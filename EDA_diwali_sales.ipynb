# **Exploratory Data Analysis**

## <font color=Black>Gender</font>

```python
sns.set(rc={'figure.figsize':(6,5)})
df = data_frame
print(df)
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
ax=sns.barplot(x='Gender', y='Amount', data=df_gender)
for bars in ax.containers:
    ax.bar_label(bars)
```

>From the above graphs most of the buyers are females and the purchasing power of females is more than males.

## Age Group

```python
ax=sns.countplot(x='Age Group', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df_age = df.groupby(['Age Group','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax=sns.barplot(x='Age Group', y='Amount', data=df_age, hue='Gender')
sns.set(rc={'figure.figsize':(15,7)})
```

> From  the above graphs it is observed that most of the buyers are of age group 26-35 females.

## State

```python
df_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,8)})
ax=sns.barplot(x='State', y='Amount', data=df_state)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,8)})
ax=sns.barplot(x='State', y='Orders', data=df_state)
for bars in ax.containers:
    ax.bar_label(bars)
```

> From the above graphs the top three states with  most of the buyers and orders are placed from Uttar Pradesh, Maharashtra and Karnataka.

  ## Occupation

```python
df_occupation=df.groupby('Occupation', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,8)})
ax=sns.barplot(x='Occupation', y='Amount', data=df_occupation)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df_occupation=df.groupby('Occupation', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,8)})
ax=sns.barplot(x='Occupation', y='Orders', data=df_occupation)
for bars in ax.containers:
    ax.bar_label(bars)
```

> From the above graphs most of the buyers are working in IT sector, Healthcare and Aviation.

## Marital_Status

```python
ax=sns.countplot(x='Marital_Status', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

> From the above graph it is observed that nost of the buyers are married(women) and have high purchasing power.


## Product_Category

```python
sns.set(rc={'figure.figsize':(20,7)})
ax=sns.countplot(x='Product_Category', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
fig1, ax1=plot.subplots(figsize=(20,7))
df.groupby('Product_Category')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
fig1, ax1=plot.subplots(figsize=(20,7))
df.groupby('Product_Category')['Amount'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

>From the above graphs it is observed that most of the sold products are from Food, Clothing and Apparel.

# Conclusion

> From the above analysis we can conclude that food, clothing and apparel products are mostly purchased by married women of age group<br> 26-35 from Uttar Pradesh, Maharashtra and Karnataka who are working in IT sector,Healthcare and Aviation.
